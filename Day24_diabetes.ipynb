```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt'
pd_data = pd.read_csv(url, sep='\t')
print(pd_data)

```

```python
_, axe =plt.subplots()
#axe.plot(pd_data['BMI'])
axe.hist(pd_data['AGE'], bins=10, edgecolor='pink', linewidth=2.0)


```

```python
_, axe =plt.subplots()
axe.boxplot(pd_data['AGE'])
```

```python
_, axe =plt.subplots()
axe.pie(pd_data['SEX'].value_counts())
```

```python
axe.set(xlim=(0,8), xticks=np.arange(1,8),
       ylim=(0,56), yticks=np.arange(0, 56, 9))

axe.set_xlim(0, 8)
axe.set_ylim(0, 56)
axe.set_xticks(np.arange(1, 8))
```
